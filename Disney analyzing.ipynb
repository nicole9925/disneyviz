```python
import pandas as pd
import numpy as np

from nltk.classify import NaiveBayesClassifier
from nltk.corpus import subjectivity
from nltk.sentiment import SentimentAnalyzer
from nltk.sentiment.util import *
from nltk.probability import FreqDist
```

```python
lyrics = pd.read_csv('final_disney_songs_cleaned.csv')
lyrics["Lyrics"] = lyrics["Lyrics"].apply(str.strip).apply(lambda x: x.replace("  ", " ").replace("'s", ""))
```

```python
remove_list = pd.read_csv("stop_words.csv")
remove_list = list(remove_list["words"])
remove_list.append("la")
remove_list.append("they're")
remove_list.append("gonna")
remove_list.append("ya")
remove_list.append("'i")
remove_list.append("hey")
remove_list.append("hey")
remove_list.append("can'twhat")
remove_list.append("yeah")
remove_list.append("'cause")
remove_list.append("ha")
remove_list.append("he'll")
remove_list.append("gotta")
remove_list.append("ooh")
remove_list.append("'em")
remove_list.append("i'd")
remove_list.append("ain't")
remove_list.append("da")
remove_list.append("till")
```

```python
def remove_stop(arr):
    arr = list(filter(None, arr))
    arr = [word for word in arr if word.strip() not in remove_list]
    return arr
```

```python
arr = (lyrics["Lyrics"]
       .replace("  ", " ")
       .str.cat(sep=" ").split(" "))
arr = list(filter(None, arr))
arr = [word for word in arr if word.strip() not in remove_list]
df = pd.DataFrame({"lyrics": arr, "count": [0]*len(arr)})
```

```python
#don't run again!!
counts = df.groupby('lyrics').count().sort_values('count', ascending=False).reset_index()
over100 = counts.loc[counts['count'] > 150]
over100["lyrics"] = over100["lyrics"].apply(lambda x: x.upper())
over100.to_csv("over100.csv", mode='a', header=True)
```

```python
relationships = lyrics.copy()
relationships["Lyrics"] = relationships["Lyrics"].apply(lambda x: np.array((x.split(" "))))
relationships ["Lyrics"] = relationships["Lyrics"].apply(stop)
```

```python
connections = pd.DataFrame(columns=["origin", "originId", "target", "targetId", "movie", "song"])
topwords = over100.reset_index()["lyrics"]
index = 0
for origin in range(len(topwords)):
    for target in range(len(topwords)):
        for row in range(relationships.shape[0]):
            if origin != target:
                if topwords[origin] in relationships.iloc[row]["Lyrics"]:
                    if topwords[target] in relationships.iloc[row]["Lyrics"]:
                        connections.loc[index] = [topwords[origin], origin, topwords[target], target, relationships.iloc[row]["Movie"], relationships.iloc[row]["Song"]]
                        index += 1
                
```

```python
connections.to_csv("connections.csv", mode='a', header=True)
```
