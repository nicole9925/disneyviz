```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
paths = []
url = "https://www.disneyclips.com/lyrics/"
page = requests.get(url)
fullSoup = BeautifulSoup(page.text, 'html.parser')
links = fullSoup.findAll('div', class_='main')[0].findAll('a')[11:]
for i in links:
    if (i.get('href') != None) and (i.get('href') != "#top"):
        paths.append(i.get('href'))
```

```python
songs = []
movies = []
lyrics = []

#individual movies
for path in paths:
    if path:
        req = requests.get('https://www.disneyclips.com/lyrics/'+path)
        soup = BeautifulSoup(req.text, 'html.parser')
        movie = soup.find('h1').text
        print(movie)
        links = soup.findAll('a')
        for i in links:
            if i.get('href') and "html" in i.get('href') and 'disneyclips.com' not in i.get('href'):
                href = i.get('href')
                movies.append(movie)
                req = requests.get("https://www.disneyclips.com/lyrics/" + href)
                minisoup = BeautifulSoup(req.text)
                lyric = minisoup.findAll('p')[1].get_text(separator=" ")
                lyrics.append(cleanhtml(lyric).lower())
                songs.append(minisoup.find('h1').text)   
```

```python
def cleanhtml(raw_html):
    cleanr = re.compile('Lyrics transcribed by Disneyclips.com')
    cleantext = re.sub(cleanr, '', raw_html)
    return cleantext
```

```python
df = pd.DataFrame({"Movie": movies, "Song": songs, "Lyrics": lyrics})
```

```python
df.to_csv('disney_songs_cleaned.csv', mode='a', header=True)
```

```python
def cleanhtml2(raw_html):
    cleanr = re.compile('\\[[\sa-z\'\&\.0-9:\-]*\\]')
    cleantext = re.sub(cleanr, '', raw_html)
    cleanr2 = re.compile('[""!?:\,\.\(\);¬°]')
    cleantext2 = re.sub(cleanr2, '', cleantext)
    cleanr3 = re.compile('\\r')
    cleantext3 = re.sub(cleanr3, '', cleantext2)
    cleanr4 = re.compile('\\n')
    cleantext4 = re.sub(cleanr4, '', cleantext3)
    cleanr5 = re.compile('\s\-\s')
    cleantext5 = re.sub(cleanr5, '', cleantext4)
    return cleantext5
```

```python
df["Lyrics"] = df["Lyrics"].apply(cleanhtml2)
df
```

```python
df.to_csv('final_disney_songs_cleaned.csv', mode='a', header=True)
```

```python
req = requests.get("https://disneymovieslist.com/list-of-disney-movies/")
```

```python
soup = BeautifulSoup(req.text)
```

```python
ls = soup.findAll("td")
count = 0
mov = []
year = []
while count < len(ls):
    mov.append(ls[count].text.replace("\xa0", ""))
    count+=1
    year.append(ls[count].text)
    count+=1
    
```

```python
years = pd.DataFrame({"movie": mov, "year": year})
```

```python
years.to_csv('years.csv', mode='a', header=True)
```
